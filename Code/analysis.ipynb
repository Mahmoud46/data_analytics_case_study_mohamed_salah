### Data Manipualtion

#### Rename Att_ to TOAtt

```python
season_2017_18.rename(columns={"Att_":"TOAtt"},inplace=True)
season_2018_19.rename(columns={"Att_":"TOAtt"},inplace=True)
season_2019_20.rename(columns={"Att_":"TOAtt"},inplace=True)
season_2020_21.rename(columns={"Att_":"TOAtt"},inplace=True)
season_2021_22.rename(columns={"Att_":"TOAtt"},inplace=True)
season_2022_23.rename(columns={"Att_":"TOAtt"},inplace=True)
season_2023_24.rename(columns={"Att_":"TOAtt"},inplace=True)
```

#### Replace 'eng Liverpool' with 'Liverpool' and 'eg Egypt' with 'Egypt' in **Squad** column values

```python
season_2017_18["Squad"]=season_2017_18["Squad"].replace("eng Liverpool","Liverpool")
season_2017_18["Squad"]=season_2017_18["Squad"].replace("eg Egypt","Egypt" )

season_2018_19["Squad"]=season_2018_19["Squad"].replace("eng Liverpool","Liverpool")
season_2018_19["Squad"]=season_2018_19["Squad"].replace("eg Egypt","Egypt" )

season_2019_20["Squad"]=season_2019_20["Squad"].replace("eng Liverpool","Liverpool")
season_2019_20["Squad"]=season_2019_20["Squad"].replace("eg Egypt","Egypt" )

season_2020_21["Squad"]=season_2020_21["Squad"].replace("eng Liverpool","Liverpool")
season_2020_21["Squad"]=season_2020_21["Squad"].replace("eg Egypt","Egypt" )

season_2021_22["Squad"]=season_2021_22["Squad"].replace("eng Liverpool","Liverpool")
season_2021_22["Squad"]=season_2021_22["Squad"].replace("eg Egypt","Egypt" )

season_2022_23["Squad"]=season_2022_23["Squad"].replace("eng Liverpool","Liverpool")
season_2022_23["Squad"]=season_2022_23["Squad"].replace("eg Egypt","Egypt" )

season_2023_24["Squad"]=season_2023_24["Squad"].replace("eng Liverpool","Liverpool")
season_2023_24["Squad"]=season_2023_24["Squad"].replace("eg Egypt","Egypt" )
```

#### Impute numerical missing values 

```python
# Imputation of Missing Values
season_attr_column_lenght=len(season_2017_18.columns)
for column in season_2017_18.columns[15:season_attr_column_lenght-1]:
    season_2017_18[column].fillna(round(season_2017_18[column].mean(),1), inplace=True)

for column in [*season_2018_19.columns][15:season_attr_column_lenght-1]:
    season_2018_19[column].fillna(round(season_2018_19[column].mean(),1), inplace=True)

for column in [*season_2019_20.columns][15:season_attr_column_lenght-1]:
    season_2019_20[column].fillna(round(season_2019_20[column].mean(),1), inplace=True)

for column in [*season_2020_21.columns][15:season_attr_column_lenght-1]:
    season_2020_21[column].fillna(round(season_2020_21[column].mean(),1), inplace=True)

for column in [*season_2021_22.columns][15:season_attr_column_lenght-1]:
    season_2021_22[column].fillna(round(season_2021_22[column].mean(),1), inplace=True)

for column in [*season_2022_23.columns][15:season_attr_column_lenght-1]:
    season_2022_23[column].fillna(round(season_2022_23[column].mean(),1), inplace=True)

for column in [*season_2023_24.columns][15:season_attr_column_lenght-1]:
    season_2023_24[column].fillna(round(season_2023_24[column].mean(),1), inplace=True)
```

### Analysis

```python
seasons_2017_24=pd.concat([season_2017_18,season_2018_19,season_2019_20,season_2020_21,season_2021_22,season_2022_23,season_2023_24])
seasons_2017_24_eg=seasons_2017_24[seasons_2017_24["Squad"]=="Egypt"]
seasons_2017_24_liv=seasons_2017_24[seasons_2017_24["Squad"]=="Liverpool"]
```

```python
def display_data_columns_sum(data_table:pd.DataFrame, squad='all'):
    '''
    squad = "liverpool" or "egypt" or "all"
    '''
    analysis_result={}
    if squad=='liverpool':
        data_table=data_table[data_table["Squad"]=="Liverpool"]
        for column in [*data_table.columns]:
            if not(any(isinstance(item, str) for item in data_table[column])):
                if column=="Cmp%":
                    analysis_result[column]=np.average(data_table[column])
                else: analysis_result[column]=sum(data_table[column])
            else: 
                analysis_result[column]='-'
                
    elif squad=='egypt':
        data_table=data_table[data_table["Squad"]=="Egypt"]
        for column in [*data_table.columns]:
            if not(any(isinstance(item, str) for item in data_table[column])):
                if column=="Cmp%":
                    analysis_result[column]=np.average(data_table[column])
                else: analysis_result[column]=sum(data_table[column])
            else: 
                analysis_result[column]='-'
    else:
        for column in [*data_table.columns]:
            if not(any(isinstance(item, str) for item in data_table[column])):
                if column=="Cmp%":
                    analysis_result[column]=np.average(data_table[column])
                else: analysis_result[column]=sum(data_table[column])
            else: 
                analysis_result[column]='-'
    return pd.DataFrame([analysis_result])
```

```python
seasons_2017_24_eg_analysis=display_data_columns_sum(seasons_2017_24_eg)
seasons_2017_24_liv_analysis=display_data_columns_sum(display_data_columns_sum(seasons_2017_24_liv))
seasons_2017_24_analysis=display_data_columns_sum(display_data_columns_sum(seasons_2017_24))
```
